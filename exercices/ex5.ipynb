```python
import numpy as np
from si.io.csv_file import read_csv
from si.decomposition.pca import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load the Iris dataset
iris = read_csv("../datasets/iris/iris.csv", features=True, label=True)

# Encode the string labels to numerical values
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(iris.y)

# Initialize PCA with 2 components
pca = PCA(n_components=2)

# Fit and transform the data
pca._fit(iris)
iris_reduced = pca._transform(iris)

# Print results
print("Original shape:", iris.X.shape)
print("Reduced shape:", iris_reduced.X.shape)
print("Explained variance:", pca.explained_variance)

# Plot the PCA results
plt.scatter(iris_reduced.X[:, 0], iris_reduced.X[:, 1], c=y_encoded, cmap='viridis')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA of Iris Dataset")
plt.colorbar(label='Class')
plt.show()
```
